Here is a simple query to find 15 runs submitted for the organsim 'Homo sapiens'.

```python
%%bigquery
SELECT *
FROM `nih-sra-datastore.sra.metadata`
WHERE organism = 'Homo sapiens'
LIMIT 15
```

If you scroll to the right you will see columns labelled attributes and jattr (attributes stored as JSON) for each row. These columns collect additional information that might not be submitted for all rows in the database. They will have a key (k) and value (v) that can be searched. In BigQuery you will do this using the UNNEST feature.

```python
%%bigquery
SELECT *
FROM `nih-sra-datastore.sra.metadata` as s
WHERE organism = 'Homo sapiens' and ( ('body_site_sam', 'peripheral blood granulocytes') in UNNEST(s.attributes) )
```

```python
%%bigquery
SELECT * 
FROM `nih-sra-datastore.sra_tax_analysis_tool.taxonomy`
WHERE sci_name = 'Homo sapiens'
```

```python
%%bigquery
SELECT a.acc, i.analyzed_spot_count, a.self_count
FROM `nih-sra-datastore.sra_tax_analysis_tool.tax_analysis` a 
    JOIN `nih-sra-datastore.sra_tax_analysis_tool.tax_analysis_info` i 
    ON a.acc = i.acc
WHERE a.tax_id = 9606
    AND a.self_count/i.analyzed_spot_count > .9
    AND i.analyzed_spot_count > 100000
```

```python
%%bigquery
SELECT m.bioproject, m.biosample, m.acc, m.collection_date_sam, m.geo_loc_name_sam, 
(select v from unnest(m.attributes) where k = 'collected_by_sam') as collected_by,
(select v from unnest(m.attributes) where k = 'host_sam') as host
FROM `nih-sra-datastore.sra.metadata` m , `nih-sra-datastore.sra_tax_analysis_tool.tax_analysis` tax
WHERE m.acc = tax.acc
and tax.name = 'Coronaviridae'
and m.geo_loc_name_sam <> ''
LIMIT 15
```
